# Example 4: Projected galaxy clustering $w_{\rm p}$

In this example, we'll show how to calculate $w_{\rm p},$ the projected clustering signal of a mock catalog.  

```python
from halotools.empirical_models import PrebuiltHodModelFactory
model = PrebuiltHodModelFactory('tinker13', threshold = 10.25)
model.populate_mock(simname = 'bolshoi', redshift = 0, halo_finder = 'rockstar')
```

Now we will extract the *x, y, z* positions of various subsamples of our galaxies 

```python
x = model.mock.galaxy_table['x']
y = model.mock.galaxy_table['y']
z = model.mock.galaxy_table['z']
```

```python
from halotools.mock_observables import return_xyz_formatted_array

all_positions = return_xyz_formatted_array(x, y, z)

red_mask = (model.mock.galaxy_table['sfr_designation'] == 'quiescent')
blue_mask = (model.mock.galaxy_table['sfr_designation'] == 'active')

red_positions = return_xyz_formatted_array(x, y, z, mask = red_mask)
blue_positions = return_xyz_formatted_array(x, y, z, mask = blue_mask)
```

```python
from halotools.mock_observables import wp, tpcf
```

```python
pi_max = 40.
rp_bins = np.logspace(-1,1.25,15)
pi_bins = np.logspace(-1,np.log10(pi_max),15)

wp_all = wp(all_positions, rp_bins, pi_bins, period=model.mock.Lbox)
wp_red = wp(red_positions, rp_bins, pi_bins, period=model.mock.Lbox)
wp_blue = wp(blue_positions, rp_bins, pi_bins, period=model.mock.Lbox)
```

```python
%matplotlib inline
from seaborn import plt
```

```python
rp_bin_centers = (rp_bins[:1] + rp_bins[1:])/2.

plt.plot(rp_bin_centers, wp_all, 
         label=r'All galaxies: $M_{\ast} > 10^{10.5}M_{\odot}$', 
         color='k')
plt.plot(rp_bin_centers, wp_red, 
         label=r'Quiescent galaxies: $M_{\ast} > 10^{10.5}M_{\odot}$', 
         color='red')
plt.plot(rp_bin_centers, wp_blue, 
         label=r'Star-forming galaxies: $M_{\ast} > 10^{10.5}M_{\odot}$', 
         color='blue')

plt.xlim(xmin = 0.1, xmax = 10)
plt.ylim(ymin = 0.5, ymax = 5e3)
plt.loglog()
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel(r'$r_{\rm p} $  $\rm{[Mpc]}$', fontsize=25)
plt.ylabel(r'$w_{\rm p}(r_{\rm p})$', fontsize=25)
plt.legend(loc='best', fontsize=20)
```

## Dissecting the impact of redshift-space distortions

The above calculation used the true spatial positions of the galaxies in the clustering calculation. Although projected clustering $w_{\rm p}$ projects out most of the impact of peculiar velocities on the measurement, the effect is certainly non-zero, especially on small scales. In this next section, we'll show how to quantitatively answer the following question: how do redshift-space distortions impact the measurement of $w_{\rm p}$?

```python
_x = np.copy(model.mock.galaxy_table['x'])
_y = np.copy(model.mock.galaxy_table['y'])
_z = np.copy(model.mock.galaxy_table['z'])



_vz = np.copy(model.mock.galaxy_table['vz'])
_halo_vz = np.copy(model.mock.galaxy_table['halo_vz'])
```

```python
true_positions = return_xyz_formatted_array(_x, _y, _z)
zspace_positions = return_xyz_formatted_array(_x, _y, _z, 
                            velocity = _vz, velocity_distortion_dimension = 'z', 
                            period = model.mock.Lbox)
halo_zspace_positions = return_xyz_formatted_array(_x, _y, _z, 
                            velocity = _halo_vz, velocity_distortion_dimension = 'z', 
                            period = model.mock.Lbox)
```

```python
xi_true = tpcf(true_positions, rp_bins, period=model.mock.Lbox)
xi_zspace = tpcf(zspace_positions, rp_bins, period=model.mock.Lbox)
xi_halo_zspace = tpcf(halo_zspace_positions, rp_bins, period=model.mock.Lbox)
```

```python
wp_true = wp(true_positions, rp_bins, pi_bins, period=model.mock.Lbox)
wp_zspace = wp(zspace_positions, rp_bins, pi_bins, period=model.mock.Lbox)
wp_halo_zspace = wp(halo_zspace_positions, rp_bins, pi_bins, period=model.mock.Lbox)
```

```python
fracdiff_wp = (wp_zspace-wp_true)/wp_true
fracdiff_wp_halo = (wp_halo_zspace-wp_true)/wp_true
fracdiff_xi = (xi_zspace-xi_true)/xi_true
fracdiff_xi_halo = (xi_halo_zspace-xi_true)/xi_true
```

```python
plt.plot(rp_bin_centers, fracdiff_wp, 
         label=r'$\mathcal{O} = w_{\rm p}; z_{\rm sat} = z_{\rm jeans}$', 
         color='green')
plt.plot(rp_bin_centers, fracdiff_wp_halo,'--',  
         label=r'$\mathcal{O} = w_{\rm p}; z_{\rm sat} = z_{\rm host-halo}$', 
         color='green')

plt.plot(rp_bin_centers, fracdiff_xi, 
         label=r'$\mathcal{O} = \xi; z_{\rm sat} = z_{\rm jeans}$', 
         color='blue')
plt.plot(rp_bin_centers, fracdiff_xi_halo, '--', 
         label=r'$\mathcal{O} = \xi; z_{\rm sat} = z_{\rm host-halo}$', 
         color='blue')

plt.xlim(xmin = 0.1, xmax = 20)
plt.xscale('log')
# plt.ylim(ymin = 0.5, ymax = 5e3)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel(r'$r $  $\rm{[Mpc]}$', fontsize=25)

plt.ylabel(r'$(\mathcal{O}_{\rm z-space} - \mathcal{O}_{\rm true}) / \mathcal{O}_{\rm true})$', fontsize=25)
plt.legend(fontsize=20, loc='best')
```

```python
halo_diff = halo_zspace_positions[:,2] - true_positions[:,2]
full_diff = zspace_positions[:,2] - true_positions[:,2]
```

```python
print(np.std(halo_diff), np.std(full_diff))
```
